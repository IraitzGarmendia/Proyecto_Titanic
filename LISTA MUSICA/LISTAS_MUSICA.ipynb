Artistas más reproducidos por semanas¶
Transforme la columna date en datetime.

Use set_index para crear un índice en la columna date.

Imprima el dataframe con el nuevo índice.

```python

# Transformamos la columna 'date' en datetime
df1['date'] = df1['date'].apply(pd.to_datetime)

# Usamos .set_index para remplazar como indice
df1.set_index('date', inplace=True)







```

Encuentre los más y menos escuchados según un intervalo de tiempo
Puede usar resample(frecuencia)[columna].

Las frecuencias que puede usar son:

H - Horaria
D - Diaria
W - Semanal
M - Mensual
A - Anual
Puede usar .agg([columnas]) para agregar los datos

```python

# Definimos la frecuencia de resampling, por ejemplo, 'D' para días
frecuencia = 'D'

# Resample los datos para sumar las veces que se ha escuchado cada canción en cada intervalo
df_resampled = df1.resample(frecuencia)['weeks-on-board'].sum()

# Ordenar las canciones más y menos escuchadas
canciones_mas_escuchadas = df_resampled.sort_values(ascending=False)
canciones_menos_escuchadas = df_resampled.sort_values()

# Mostrar las canciones más escuchadas
print("Las canciones más escuchadas son:")
print(canciones_mas_escuchadas)

# Mostrar las canciones menos escuchadas
print("\nLas canciones menos escuchadas son:")
print(canciones_menos_escuchadas)
```
